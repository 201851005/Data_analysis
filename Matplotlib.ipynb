```python
# import the library
import matplotlib.pyplot as plt
```

```python
plt.plot([1,3,7,3])  # y values is [1,3,7,3] and x values is [0,1,2,3]
plt.xlabel("indecies")
plt.ylabel("values")
plt.title("myPlot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.title("SquarePlot")
plt.grid() #on grid
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y,'ro')  # r-> red ,o ->dots
plt.title("SquarePlot")
plt.grid() #on grid
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0.,5.,0.2)


fig1 = plt.figure(figsize =(10,5))

# blue dashes, red square, green triangle
plt.plot(t,t**2, 'b--', label = '^2')
plt.plot(t,t**2.2, 'rs', label = '^2.2')
plt.plot(t,t**2.5, 'g^', label = '^2.5')
plt.legend() # add legend
plt.grid()
plt.show()

```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y,linewidth = 5)  # use linewidth
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1,y1,x2,y2)  # lines[0] plot by x1,y1  and lines[1] plot by x2,y2.

plt.setp(lines[0], color ='r', linewidth =1.4)  # setp --> set properties

plt.setp(lines[1], 'color', 'g', linewidth =1.9)
plt.grid()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

plt.figure(1)

plt.subplot(211)  # 2 is number of rows, 1 is column , 1 for the position
plt.grid()
plt.plot(x1,y1,'g--')

plt.figure(212)
plt.grid()
plt.plot(x2,y2,'b^')
plt.show()
```

```python
plt.figure(1)

plt.subplot(211)
plt.plot([1,2,3,4,5], 'r--')
plt.grid()

plt.subplot(212)
plt.plot([11,15,7,8], 'g--')
plt.grid()

plt.figure(2)

plt.subplot(211)
plt.plot([2,4,9,16,25], 'b--')
plt.grid()
plt.show()
```

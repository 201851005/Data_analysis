```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

x = stats.norm.rvs(size =1000)  # generate 1000 points for normal distribution
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw =0.5)  # kernel-density plot 
plt.show()
```

```python
stats.kstest(x,'norm')
```

pvalue is quite high, this means distribution are same


```python
# now generate sample for uniform distribution

y = np.random.uniform(0,1,10000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y,'norm')
```

pvalue is 0 , which means distribution are not same

```python
stats.kstest(y,'uniform')
```

```python
import numpy as np
import pylab
import scipy.stats as stats 

# N(0,1)

std_normal = np.random.normal(loc = 0, scale = 1, size = 1000)

for i in range(0,101):
    print(str(i)+"-> "+ str(np.percentile(std_normal,i)))
```

```python
# generate 100 samples from N(20,5)
measurement = np.random.normal(loc = 20, scale = 5, size=100)
# try size 1000

stats.probplot(measurement, dist ="norm",plot = pylab)
pylab.show()
```

```python
measurement = np.random.normal(loc = 20, scale = 5, size=1000)
# try size 1000

stats.probplot(measurement, dist ="norm",plot = pylab)
pylab.show()
```

```python
# lets generate 1000 sample for unkiform distribution
measurement = np.random.uniform(low = -1, high = 1, size=1000)
# try size 1000

stats.probplot(measurement, dist ="norm",plot = pylab)
pylab.show()
```
